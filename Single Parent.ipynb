```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#assumptions
inf_av = 0.95
inf_s  = 0.5

#first generation population
n = 100

#new generation
n_next  = 0
inf_num = 0

for i in range(n):
    inf_num = int(np.random.normal(inf_av, inf_s, 1))
    n_next = n_next + inf_num
            
print(n_next)
```

```python
#gaussian distribution assumption

def next_gen(n, inf_av, inf_s):
    n_next  = 0
    inf_num = 0

    for i in range(n):
        inf_num = int(np.random.normal(inf_av, inf_s, 1))
        n_next += inf_num
    return n_next
```

```python
next_gen(n, inf_av, inf_s)
```

```python
x = 20
T = np.arange(x)
pop = np.zeros(x)
pop[0] = 10000

ave = 100
ns = np.zeros(ave)


for t in range((x-1)):
    for i in range(ave):
        ns[i] = next_gen(int(pop[t]), inf_av, inf_s)
    pop[t+1] = np.mean(ns)
    
plt.plot(pop, T)
plt.title("PP for {} generation averaged on {} samples for {} as initial population".format(x, ave, pop[0]))
```

```python
%%time

for inf_av in [0.8, 1, 1.1, 1.3, 1.5, 2]:
    x = 10
    T = np.arange(x)
    pop = np.zeros(x)
    pop[0] = 1000

    ave = 30
    ns = np.zeros(ave)


    for t in range((x-1)):
        for i in range(ave):
            ns[i] = next_gen(int(pop[t]), inf_av, inf_s)
        pop[t+1] = np.mean(ns)

    plt.plot(pop, T, label = 'Fertility = {}'.format(inf_av))
    plt.title("PP for {} generation averaged on {} samples for {} as initial population".format(x, ave, pop[0]))
    
plt.legend()
```
