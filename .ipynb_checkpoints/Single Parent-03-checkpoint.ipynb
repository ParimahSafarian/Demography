```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

### Poisson distribution for number of childeren
The distrbution is discrete. It has been used for this purpose and its only parameter $\lambda$ makes sense!

```python
#Poisson distribution assumption

def poisson_next_gen(n, lam):
    n_next  = 0
    inf_num = 0

    for i in range(n):
        inf_num = int(np.random.poisson(lam, 1))
        n_next += inf_num
        #need to assume logic range
#         if inf_num < 10:
#             n_next +=inf_num
    return n_next
```

```python
%%time

lam_mat = np.arange(0.9, 1.1, 0.02)

 

for lam in lam_mat:
    generation   = 10
    time    = np.arange(generation)
    pop     = np.zeros(generation)
    pop[0]  = 100
    pop_std = np.zeros(generation)

    ave = 200
    ns = np.zeros(ave)


    for t in range((generation-1)):
        for i in range(ave):
            ns[i] = poisson_next_gen(int(pop[t]), lam)
        pop[t+1]     = np.mean(ns)
        pop_std[t+1] = np.std(ns) 

    plt.plot(pop, time, label = '$\lambda = ${}'.format(np.round(lam, 3)))
    plt.title("PP for {} generation averaged on {} samples for {} as initial population \
    \n Using poisson distribution".format(generation, ave, pop[0]))
    
#     plt.plot(pop_std, time, label = f'std for $\lambda$ = {lam}')
    
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

### Special Distribution

```python
#make a distribution function

def inf_dist(prob):
    rand = random.random() * 100
    if   rand <= prob[0]:
        return 0
    elif rand <= prob[1]:
        return 1
    elif rand <= prob[2]:
        return 2
    elif rand <= prob[3]:
        return 3
    
    
#p in an array which presents the probability to have i childeren
#5 is the upper limit
```

Using Iran's data for household size prob must be like:

```python
prob = [(8.5 + 20.7), (8.5+20.7+28.5), (8.5+20.7+28.5+27.6), (8.5+20.7+28.5+27.6+14.7)]
prob
```

```python
def arbitrary_next_gen(n, prob):
    n_next  = 0
    inf_num = 0

    for i in range(n):
        inf_num = inf_dist(prob)
        n_next = n_next + inf_num
    return int(n_next)
```

```python
arbitrary_next_gen(100, prob)
```

```python
generation = 10
time   = np.arange(generation)
pop    = np.zeros(generation)
pop[0] = 100

ave = 100
ns = np.zeros(ave)

for t in range((generation-1)):
    for i in range(ave):
        ns[i] = arbitrary_next_gen(int(pop[t]), prob)
    pop[t+1] = np.mean(ns)
    
plt.plot(pop, time)
plt.title("PP for {} generation averaged on {} samples for {} as initial population \
\ \n Using Iran 1395 distribution\n NOTE: This is a single parent model".format(generation, ave, pop[0]))
```

```python
%%time

lam_mat = np.arange(0.9, 1.1, 0.02)
for lam in lam_mat:
    generation   = 50
    time   = np.arange(generation)
    pop    = np.zeros(generation)
    pop[0] = 50

    ave = 200
    ns = np.zeros(ave)


    for t in range((generation-1)):
        for i in range(ave):
            ns[i] = poisson_next_gen(int(pop[t]), lam)
        pop[t+1] = np.mean(ns)

    plt.plot(pop, time, label = '$\lambda = ${}'.format(np.round(lam, 3)))
    plt.title("PP for {} generation averaged on {} samples for {} as initial population\
    \ \n Using poisson distribution".format(generation, ave, pop[0]))
    
plt.legend()
```

### Extinction time

```python
%%time

lam_mat = np.arange(0, 1.2, 0.05)
t_c  = np.zeros(np.size(lam_mat))

for m in range(np.size(lam_mat)):
    lam = lam_mat[m]
    generation = 50
    time = np.arange(generation)
    pop  = np.zeros(generation)
    
    pop[0] = 50

    ave = 50
    ns = np.zeros(ave)

    for t in range((generation-1)):
        if pop[t] >= pop[0] * 1.1:
            t_c[m] = generation
            break
        for i in range(ave):
            ns[i] = poisson_next_gen(int(pop[t]), lam)
            
        pop[t+1] = np.mean(ns)
        if pop[t+1] <= 10 :
            t_c[m] = (t+1)
            break
            
    plt.plot(pop, time, label = 'Fertility = {}'.format(lam))
    plt.title("PP for {} generation averaged on {} samples for {} as initial population".format(generation, ave, pop[0]))
print(t_c)

# plt.plot(lam_mat, t_c)
```

```python
%%time

lam_mat = np.arange(0, 1.1, 0.005)
t_c  = np.zeros(np.size(lam_mat))

for m in range(np.size(lam_mat)):
    lam = lam_mat[m]
    generation = 50
    time = np.arange(generation)
    pop  = np.zeros(generation)
    
    pop[0] = 50

    ave = 50
    ns = np.zeros(ave)

    for t in range((generation-1)):
        if pop[t] >= pop[0] * 1.1:
            t_c[m] = generation
            break
        for i in range(ave):
            ns[i] = poisson_next_gen(int(pop[t]), lam)
            
        pop[t+1] = np.mean(ns)
        if pop[t+1] <= 10 :
            t_c[m] = (t+1)
            break
            
#     plt.plot(pop, time, label = 'Fertility = {}'.format(lam))


plt.plot(lam_mat, t_c, c = 'indigo')
plt.title("Extinction time versus avarage number of enfants\n \
\ using poisson distribution")
plt.ylabel("$t_c$")
plt.xlabel("$\lambda$")



print(lam_mat, t_c)
```
